### Configuring a `consumer that subscribes` (read and process) records from Kafka
####  `Consumer` consumes records from a Kafka Cluster:

```python
def CreateKafkaConsumer():
    topic_name = 'SmokeDetection' # topic where the messages are  published
    localhost = 'localhost:9092'
    return KafkaConsumer(topic_name, bootstrap_servers=[localhost],value_deserializer=lambda x:loads(x.decode('utf-8')))
```

### Printing `records` that are consumed from a Kafka Cluster

```python
def RecordsSSDConsumedFromKafka():
    consumer = CreateKafkaConsumer()
    for message in consumer:
        print(message.value)

RecordsSSDConsumedFromKafka()
```
